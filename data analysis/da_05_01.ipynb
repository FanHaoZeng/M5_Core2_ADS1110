```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 读取数据文件
df = pd.read_csv('../data/ads_20250501_160504_749.csv')


# 重命名列名
df.columns = ['timestamp', 'voltage']

# 将时间戳转换为datetime类型
df['timestamp'] = pd.to_datetime(df['timestamp'])



# 显示数据基本信息
print("数据基本信息：")
print(df.info())

# 显示前几行数据
print("\n数据预览：")
display(df.head())
```

```python
plt.figure(figsize=(12, 6))
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.plot(df['timestamp'], df[column], label=column)
plt.legend()
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 计算前10%的数据点数量
n_points = int(len(df) * 0.1)

# 绘制前10%的数据
plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'][:n_points], df['voltage'][:n_points], label='Voltage')
plt.title('First 10% of Voltage Data')
plt.xlabel('Timestamp')
plt.ylabel('Voltage (mV)')
plt.grid(True)
plt.legend()
plt.show()
```
