```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False  
```

# ads_20250511_184735_675.csv

```python
# 读取数据文件
df = pd.read_csv('../data/ads_20250511_184735_675.csv')


# 重命名列名
df.columns = ['timestamp','voltage', 'duration']

# 将时间戳转换为datetime类型
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 显示数据基本信息
print("数据基本信息：")
print(df.info())

# 显示前几行数据
print("\n数据预览：")
display(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['duration'] / 60000, df['voltage'], label='电压变化', color='b')  # 将ms转换为分钟
plt.xlabel('Duration (minutes)')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Convert specific time point to datetime format
specific_time = pd.to_datetime('2025-05-11T18:54:26.821Z')

# Check if this time point is within the data range
if specific_time >= df['timestamp'].min() and specific_time <= df['timestamp'].max():
    # Find the closest time point
    closest_idx = (df['timestamp'] - specific_time).abs().idxmin()
    closest_point = df.iloc[closest_idx]
    
    # Calculate the time window (5 minutes before and after the specific time)
    closest_duration_min = closest_point['duration'] / 60000  # Convert to minutes
    time_window_min = 3  # 5 minutes window
    
    # Filter data within the time window
    min_duration = (closest_duration_min - (time_window_min - 1.7)) * 60000
    max_duration = (closest_duration_min + (time_window_min + 1.4)) * 60000
    filtered_df = df[(df['duration'] >= min_duration) & (df['duration'] <= max_duration)]
    
    # Plot voltage changes over time, and mark the specific time point
    plt.figure(figsize=(18, 6))
    plt.plot(filtered_df['duration'] / 60000, filtered_df['voltage'], label='Voltage Change', color='b')
    
    # Mark the specific time point
    plt.scatter(closest_point['duration'] / 60000, closest_point['voltage'], 
                color='red', s=100, zorder=5)
    plt.annotate(f'1200Hz Sine Wave 10s, \nVoltage: {closest_point["voltage"]:.2f}mV', 
                xy=(closest_point['duration'] / 60000, closest_point['voltage']),
                xytext=(closest_point['duration'] / 60000 - 0.5, closest_point['voltage'] + 0.1),
                fontsize=10)
    
    # Set axis labels and title
    plt.xlabel('Time (min)')
    plt.ylabel('Voltage (mV)')
    plt.title(f'Voltage Data Around sound situmilation')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("The specified time point is not within the data range!")
    print(f"Data time range: {df['timestamp'].min()} to {df['timestamp'].max()}")
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from datetime import datetime

# 定义一个函数来读取CSV文件并处理数据
def read_ads_csv(file_path):
    try:
        # 读取CSV文件
        df = pd.read_csv(file_path, header=None, names=['timestamp', 'voltage', 'duration'])
        
        # 将时间戳转换为datetime格式
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        
        return df
    except Exception as e:
        print(f"处理文件 {file_path} 时出错: {e}")
        return None

# 指定要处理的文件列表
file_list = [
    '../data/ads_20250511_203720_944.csv',
    '../data/ads_20250511_204305_785.csv',
    '../data/ads_20250511_204745_713.csv',
    '../data/ads_20250511_204930_339.csv',
    '../data/ads_20250511_205844_068.csv',
    '../data/ads_20250511_210500_341.csv',
    '../data/ads_20250511_212117_313.csv'
]

# 读取所有文件并合并数据
dfs = []
for file_path in file_list:
    df = read_ads_csv(file_path)
    if df is not None:
        dfs.append(df)

# 检查是否有成功读取的数据
if not dfs:
    print("没有成功读取任何数据文件")
else:
    # 合并所有数据帧
    combined_df = pd.concat(dfs, ignore_index=True)
    
    # 按时间戳排序
    combined_df = combined_df.sort_values('timestamp')
    
    # 重新计算duration，基于排序后的timestamp
    # 将timestamp转换为毫秒级时间戳
    combined_df['duration'] = (combined_df['timestamp'] - combined_df['timestamp'].min()).dt.total_seconds() * 1000
    
    # 显示合并后的数据
    print("合并后的数据（前5行）:")
    print(combined_df.head())
    
    # 绘制电压随时间变化的图表
    plt.figure(figsize=(18, 6))
    plt.plot(combined_df['duration'] / 60000, combined_df['voltage'], label='电压变化')
    
    plt.xlabel('时间（分钟）')
    plt.ylabel('电压（mV）')
    plt.title('合并的电压数据')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # 将结果保存在df变量中
    df = combined_df.copy()
    print(f"合并的数据已保存到变量 df 中，共 {len(df)} 行")
```

```python
# 绘制电压随实际时间变化的图表
plt.figure(figsize=(18, 6))
plt.plot(combined_df['timestamp'], combined_df['voltage'], label='电压变化')

plt.xlabel('时间')
plt.ylabel('电压（mV）')
plt.title('合并的电压数据（实际时间）')
plt.legend()
plt.grid(True)
plt.gcf().autofmt_xdate()  # 自动格式化x轴日期标签
plt.tight_layout()
plt.show()
```
