```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 读取数据文件
df = pd.read_csv('../data/ads_20250503_180543_869.csv')

# 删除最后一列
df = df.iloc[:, :-1]

# 重命名列名
df.columns = ['timestamp', 'voltage']

# 将时间戳转换为datetime类型
df['timestamp'] = pd.to_datetime(df['timestamp'])



# 显示数据基本信息
print("数据基本信息：")
print(df.info())

# 显示前几行数据
print("\n数据预览：")
display(df.head())
```

```python
plt.figure(figsize=(12, 6))
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.plot(df['timestamp'], df[column], label=column)
plt.legend()
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 计算最后10%的数据点数量
last_10_percent = int(len(df) * 0.0001)

# 获取最后10%的数据
last_10_percent_data = df.tail(last_10_percent)

# 显示最后10%的数据
print("最后10%的数据：")
display(last_10_percent_data)

# 绘制最后10%数据的时序图
plt.figure(figsize=(12, 6))
plt.scatter(last_10_percent_data['timestamp'], last_10_percent_data['voltage'], label='voltage')
plt.legend()
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Last 10% of Time Series Data')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
