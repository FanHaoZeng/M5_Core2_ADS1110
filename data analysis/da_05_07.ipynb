```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 读取数据文件
df = pd.read_csv('../data/ads_20250507_163940_306.csv')


# 重命名列名
df.columns = ['timestamp','voltage', 'duration']

# 将时间戳转换为datetime类型
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 显示数据基本信息
print("数据基本信息：")
print(df.info())

# 显示前几行数据
print("\n数据预览：")
display(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['duration'] / 60000, df['voltage'], label='电压变化', color='b')  # 将ms转换为分钟
plt.xlabel('Duration (minutes)')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 使用真实时间作为x轴
plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['voltage'], label='电压变化', color='b')
plt.xlabel('时间')
plt.ylabel('电压 (mV)')
plt.title('电压随时间变化曲线')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # 旋转x轴标签以便更好地显示时间
plt.tight_layout()  # 自动调整图表布局
plt.show()
```
